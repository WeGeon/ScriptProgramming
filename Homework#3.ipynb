### 4. 다음 문자열을 ' : ' 기준으로 분리하여 리스트로 만들고 각 문자열의 좌우 공백을 제거하시오 ( S -> I)
> s = '&nbsp;&nbsp;first star&nbsp;&nbsp;&nbsp;:&nbsp;&nbsp;&nbsp;second star&nbsp;&nbsp;&nbsp;:&nbsp;&nbsp;&nbsp;&nbsp;third star&nbsp;&nbsp;'  
> I = ['first star', 'second star', 'third star']   
>  
>- 1) for ~ in 문을 이용한 코드를 작성하시오.  
>- 2) 리스트 내포를 이용한 코드를 작성하시오.

```python
s = '  first star   :   second star   :    third star  '
mid = s.split(':')
l = [i.strip() for i in mid]
print("s = " + s)
print("l = {0}".format(l))
```

#### <결과>
> s 문자열에서 split을 통해 ':'을 기준으로 나눈 문자열을 mid라는 list로 변환한 뒤,  
> for i in mid를 통하여 mid안의 각 인덱스 값들을 순서대로 i가 가리키게 한 뒤, 그 값을 .strip()으로 양 공백을 제거한다.  
>- 1) for ~ in 문을 이용한 코드를 작성하시오.  
>> for i in mid  => mid 안에 있는 각 요소들을 i가 가리키게 한다.
>- 2) 리스트 내포를 이용한 코드를 작성하시오.  
>> l = [i.strip() for i in mid] => 반복을 통해 생성되는 i의 양 공백을 제거한 뒤 l 리스트에 삽입한다.

### 5. 다음과 같이 0보다 큰 정수 리스트 변수 list를 인자로 받는 함수 addall(list)와 addallodd(list)를 for ~ in 리터럴과 리스트 내포 방식으로 각각 제시하시오.
>- 1) addall(list) 함수 (리스트 내의 모든 정수를 더하는 함수로서 해답으로는 for ~ in 리터럴과 리스트 내포 방식으로 각각 제시하시오.)
>> - addall([1])  
>> 1  
>> - addall([1, 2, 3, 4, 5, 6, 7, 8, 9])  
>> 45  
>- 2) addallodd(list) 함수 (리스트 내의 모든 홀수를 더하는 함수로서 해답으로는 for ~ in 리터럴과 리스트 내포 방식으로 각각 제시하시오.)  
>> - addallodd([1])  
>> 1  
>> - addallodd([1, 2, 3, 4, 5, 6, 7, 8, 9])  
>> 25  

>- [참고 1]:리스트 내포 방식으로 addall() 및 addalldoo()함수를 만들 때엔 리스트 내포에 의해 생성된 리스트 결과에 대해 내장함수를 사용하는 방식을 고려해보시오. (주의 : 리스트 내포 방식의 addall() 및 addallodd() 함수의 몸체는 단 1줄로 작성되어야 한다.)  
>- [참고 2]: sum() 등의 내장함수 사용 가능함  

```python
# for ~ in 리터럴
## addall()
def addall(list):
    sum = 0
    for i in list:
        if i > 0: sum += i
    return sum

## addallodd()
def addallodd(list):
    sum = 0
    for i in list:
        if i > 0 and i % 2 : sum += i
    return sum
list1 = [1]
listMany = [1, 2, 3, 4, 5, 6, 7, 8, 9]

print("1) for ~ in 리터럴")
print("addall() 함수")
print(addall(list1))
print(addall(listMany))
print("addallodd() 함수")
print(addallodd(list1))
print(addallodd(listMany))

# 리스트 내포
## addall()
def addall(list):
    return sum([i for i in list if i > 0])
## addallodd()
def addallodd(list):
    return sum([i for i in list if i > 0 and i % 2])

print("\n2) 리스트 내포")
print("addall() 함수")
print(addall(list1))
print(addall(listMany))
print("addallodd() 함수")
print(addallodd(list1))
print(addallodd(listMany))
```

#### <결과>
>- 1) for ~ in 리터럴 함수    
>>- addall(list) 함수  
>> 합을 저장할 변수 sum을 0으로 초기화 한 후,  
>> for i in list:를 통하여 i에 list 요소들을 순서대로 대입한다.  
>> 그 후, if i > 0이라면 sum에 i를 추가한다.  
>>  &nbsp;
>>- addallodd(list) 함수  
>> addall(list) 함수에서 조건에 i % 2 즉, 2로 나누어 떨어지지 않을 때를 추가한다.

>- 2) 리스트 내포 함수  
>>- addall(list) 함수  
>> result = [i for i in list if i > 0]  
>> list 안의 요소를 i에 대입 후 i가 0보다 크다면 result 안에 i를 삽입한다.  
>> 그 후, return sum(result)를 통하여 result 리스트 안의 모든 요소를 더한 값을 반환한다.  
>>  &nbsp;
>>- addallodd(list) 함수  
>> addall(list) 함수에서 조건에 if i % 2를 추가하여 홀수일 때의 i를 삽입한다.

>- 1) 위 코드의 수행 결과 d와 e는 항상 같은 결과를 출력한다. 그 이유는 무엇인가?  
>- 2) 위 코드의 수행 결과 마지막 f의 내용은 다음과 같다. 이 때 'high' 키의 값 내용 중 500이 출력된 이유와 'low'키의 값 내용 중 [1, 2, 3]이 출력된 이유를 설명하시오 ([hint] : shallow copy 관점에서 설명하시오.)  
>  {'high' : [4, 500, 6], 'low', [1, 2, 3]}

```python
L1 = [1, 2, 3]
L2 = [4, 5, 6]
d = {'low':L1, 'high':L2}
e = d
f = d.copy()
print(d)
print(e)
print(f)
print()
d['low'] = [10, 20, 30]
d['high'][1] = 500
print(d)
print(e)
print(f)
print()

print(id(d['low']))
print(id(L1))
print(id(f['high']))
print(id(L2))
```

##### <풀이> 
>- 1) d와 e는 얕은 복사로 같은 주소값을 참조하기 때문인다.  
>- 2 - 1) 'high' 키의 경우 d['high'][1] = 500을 통해 d['high'] 즉, L2의 주소에서 [1]째의 값을 직접 500으로 바꾸었기에,  
> &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; L2를 가리키는 f의 'high' 또한 [4, 500, 6]으로 바뀌었다.
>- 2 - 2) 'low' 키의 경우 d['low']가 가리키는 리스트의 주소가 [1, 20, 30]의 1980737436608로 바뀌었지만, f의 경우는 그대로 L1의  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; 1980735556864을 가리키고 있기에 변하지 않는다.  
* 즉 d와 e의 경우같은 얕은 복사는 같은 대상의 주소를 참조하기 때문에, 한쪽에 변화가 있을 시, 다른 쪽도 똑같이 변하지만, d와 f와 같은 깊은 복사의 경우는 새로운 객체를 만들어서 참조하고, 내부 요소를 직접적으로 바꾸지 않는 한 변화가 없다.

>- **Strange**라는 숫자를 아래와 같이 정의한다.  
>>- 음이 아닌 정수 중 자릿수가 1인 모든 숫자는 **Strange**이다.
>>- **length(X)** > 1인 숫자 X는 다음 2가지를 만족하면 **Strange**이다.
>>>- X는 **length(X)** 로 정확히 나누어 떨어진다.
>>>- X/length(X)가 **Strange**이다.

> 이 문제에서 숫자하나가 주어질 때, 이 숫자가 Strange인지 판단하는 함수 isStrange(num)을 작성하시오.  
> isStrange(num) 함수가 재귀적인 함수가 되도록 작성해보세요.  

```python
# 10진수로 표현된 X의 자릿수
def length(num):
    return len(str(num))

# 주어진 숫자가 Strange인지 판단하는 함수
def isStrange(num):
    if num >= 0:
        if length(num) == 1:
            return True
        elif length(num) > 1 and num % length(num) == 0:
            return isStrange(num // length(num))    
    return False

print(isStrange(0))
print(isStrange(56))
print(isStrange(100))
print(isStrange(7882602506492313600))
print(isStrange(6012302549522098651))
```

>- def isStrange(num) 함수는 입력 받은 숫자 num이 0을 포함한 양의 정수일 때,
>> 자릿수가 1인 수라면 True  
>> 또, 2 이상의 자릿수를 가지며, length(num)으로 나누어 떨어지는 수일 때, isStrange(num//length(num))을 통해 몫을 정수로 받는 재귀함수를 활용하여 1의 자릿수까지 간다면 True, 도중에 조건에 맞지 않는다면 False를 반환한다.  

> 음수, 조건이 안맞으면 False를 반환한다.

또한, 이들은 그룹의 번호순으로 여행을 하길 원한다.  
예를 들어, 3번 그룹은 1번 그룹과 2번 그룹이 여행을 한 뒤에 버스를 탈 수 있거나, 1번 그룹과 2번 그룹이 이미 버스에 탑승한 상태에 있을 때 동시에 버스를 탈 수 있다.  
마지막으로, 버스는 빈 좌석이 없을 때에만 여행을 떠난다.  

>- 이 문제에서 각 그룹의 팀원 수가 리스트로 주어질 때, 가능한 모든 버스의 사이즈 **x**(x명의 사람들이 동시에 탈 수 있음)를 반환하는 함수 solve(a) 를 작성하시오.

>- [주의] 반드시 계산 복잡도를 최대한 낮추세요. 아래 예시 3번을 실행했을 때 결과가 바로 나타날 수 있도록!!

```python
import time

# 약수를 구하는 함수
def getDivisor(n):
    d = []
    for i in range(1, int(n**(1/2)) + 1):
        if (n % i == 0):
            d.append(i) 
            if ( (i**2) != n) : 
                d.append(n // i)
    d.sort()
    return d


# 버스의 사이즈를 반환하는 함수
def solve(a):
    result = []
    sumL = 0
    copyList = []
    max = list(set(a))[-1]
    div = getDivisor(sum(a))
    for i in div:
        if i >= max:
            copyList.append(i)
    for i in copyList:
        count = 1
        for j in a:
            sumL += j
            if sumL == i:
                sumL = 0
            elif sumL > i:
                count = 0
                break
        if count == 1:
            result.append(i)
    return result

a = [1, 2, 1, 1, 1, 2, 1, 3]
b = [1, 1, 1, 1]
c = [9999]*10**5
print(solve(a))
print(solve(b))
start = time.time()
print(solve(c))
end = time.time()
print('소요된 시간 : ', end - start)
```

```python
import requests  # HTTP 요청 처리를 위한 requests 호출
import string    # 구두문자를 제거하기 위한 string 호출

# 엔티티 및, 이스케이프 시퀀스 제거 함수
def deleteStrinExEn(source):
    # html 엔티티값 제거
    source = source.replace("&nbsp;", "")
    source = source.replace("&lt;", "")
    source = source.replace("&gt;", "")
    source = source.replace("&amp;", "")
    source = source.replace("&apos;", "")
    source = source.replace("&quot;", "")
    source = source.replace("&lsquo;", "")
    source = source.replace("&rsquo;", "")
    source = source.replace("&ldquo;", "")
    source = source.replace("&rdquo;", "")
    source = source.replace("&deg;", "")
    source = source.replace("&middot;", "")
    source = source.replace("&times;", "")
    source = source.replace("&divide;", "")
    source = source.replace("&minus;", "")
    source = source.replace("&infin;", "")
    source = source.replace("&sim;", "")
    source = source.replace("&ne;", "")
    source = source.replace("&le;", "")
    source = source.replace("&ge;", "")
    source = source.replace("&ndash;", "")
    source = source.replace("&mdash;", "")
    source = source.replace("&bull;", "")
    source = source.replace("&hellip;", "")
    source = source.replace("&prime;", "")
    source = source.replace("&prime;", "")
    source = source.replace("&copy;", "")
    source = source.replace("&reg;", "")
    source = source.replace("&trade;", "")
    
    # html 이스케이프 시퀀스 제거
    source = source.replace("\n", "")  # source_fow 문자열 내에 '\n' 문자를 ''로 대체 (삭제)
    source = source.replace("\t", "")  # source_fow 문자열 내에 '\t' 문자를 ''로 대체 (삭제)
    source = source.replace("\r", "")  # source_fow 문자열 내에 '\r' 문자를 ''로 대체 (삭제)
    
    return source


# html 태그 제거 함수
def print_string(str1):  # str1 문자열을 인자로 받아옴
    result = []  # 반환할 문자열을 담을 리스트 result 생성
    while True:  # 모든 문자열을 추출할 때까지 반복
        i = 0  # 반복될때마다 인덱스 i = 0 으로 초기화
        # <script> 태그 제거
        if str1.find('<script>') == i:  # 문자열 중 <script>로 시작하는 인덱스가 시작과 같다면
            i = str1.find('</script>') + 9  # </script>가 닫히는 인덱스로 간 뒤 그 뒤에 인덱스 i에 저장
            str1 = str1[i:]  # i 인덱스 뒤로 슬라이싱 후 대입
            i = 0  # i = 다시 시작위치인 0으로 초기화

        # <script ~ > 태그 제거
        elif str1.find('<script ') == i:  # 문자열 중 <scirpt ~ >로 시작하는 인덱스가 시작과 같다면
            i = str1.find('</script>') + 9  # </script>가 닫히는 인덱스로 간 뒤 그 뒤에 인덱스 i에 저장
            str1 = str1[i:]  # i 인덱스 뒤로 슬라이싱 후 대입
            i = 0  # i = 다시 시작위치인 0으로 초기화

        # <SCRIPT> 태그 제거
        elif str1.find('<SCRIPT>') == i:  # 문자열 중 <SCRIPT>로 시작하는 인덱스가 시작과 같다면
            i = str1.find('</SCRIPT>') + 9  # </SCRIPT>가 닫히는 인덱스로 간 뒤 그 뒤에 인덱스 i에 저장
            str1 = str1[i:]  # i 인덱스 뒤로 슬라이싱 후 대입
            i = 0  # i = 다시 시작위치인 0으로 초기화

        # <style> 태그 제거
        elif str1.find('<style') == i:  # 문자열 중 <style>로 시작하는 인덱스가 시작과 같다면
            i = str1.find('</style>') + 8  # </style>이 닫히는 인덱스로 간 뒤 그 뒤에 인덱스 i에 저장
            str1 = str1[i:]  # i 인덱스 뒤로 슬라이싱 후 대입
            i = 0  # i = 다시 시작위치인 0으로 초기화

        # <STYLE> 태그 제거
        elif str1.find('<STYLE>') == i:  # 문자열 중 <STYLE>로 시작하는 인덱스가 시작과 같다면
            i = str1.find('</STYLE>') + 8  # </STYLE>이 닫히는 인덱스로 간 뒤 그 뒤에 인덱스 i에 저장
            str1 = str1[i:]  # i 인덱스 뒤로 슬라이싱 후 대입
            i = 0  # i = 다시 시작위치인 0으로 초기화

        # <SPAN ~ > 태그 제거
        elif str1.find('<SPAN') == i:  # 문자열 중 <SPAN ~ >로 시작하는 인덱스가 시작과 같다면
            i = str1.find('</SPAN>') + 7  # </SPAN>이 닫히는 인덱스로 간 뒤 그 뒤에 인덱스 i에 저장
            str1 = str1[i:]  # i 인덱스 뒤로 슬라이싱 후 대입
            i = 0  # i = 다시 시작위치인 0으로 초기화

        # <!-- --> 주석태그 제거
        elif str1.find('<!--') == i:  # 문자열 중 <SPAN ~ >로 시작하는 인덱스가 시작과 같다면
            i = str1.find('-->') + 3  # </SPAN>이 닫히는 인덱스로 간 뒤 그 뒤에 인덱스 i에 저장
            str1 = str1[i:]  # i 인덱스 뒤로 슬라이싱 후 대입
            i = 0  # i = 다시 시작위치인 0으로 초기화

        # 나머지 '<'' 로 시작하는 모든 태그 제거
        elif str1[i] == '<':  # 문자열 중 '<'로 시작하는 인덱스가 시작과 같다면
            while True:  # < > 내부의 중복 <>처리를 위한 반복
                i += 1  # 첫 시작 : '<' 다음, 반복시 : '>' 다음 인덱스
                if '<' in str1[i:str1.find('>') + 1]:  # <> 내부에 '<'가 있다면
                    i = str1.find('>') + 1  # 처음 '>'가 오는 인덱스 다음으로 인덱스 i 지정
                    str1 = str1[i:]  # i부터 시작하여 문자열 슬라이싱
                    i = 0  # i = 0으로 초기화
                else:  # '<' 가 중복으로 들어있지 않다면
                    i = str1.find('>') + 1  # '>'로 닫히는 인덱스로 간 뒤 그 뒤에 인덱스를 i에 저장
                    str1 = str1[i:]  # i 인덱스 뒤로 슬라이싱 후 대입
                    break  # '<' 내부에 <>가 있는지 판별하는 반복 탈출

        # 글자 또는 공백일 때
        else:  # '<'가 포함된 태그로 시작하지 않는다면
            if str1[i] == ' ':  # 공백이라면
                i += 1  # 다음 인덱스 값을 i에 저장 후
                str1 = str1[i:]  # i를 시작으로 전체 슬라이싱 후 대입
            else:  # 문자라면
                result.append(str1[i:str1.find('<')])  # 다음 태그가 시작되기 전까지 result에 문자열 대입
                i = str1.find('<')  # '<' 다음 인덱스 값을 i에 저장 후
                str1 = str1[i:]  # 다음 태그가 시작되는 인덱스부터 시작
        if len(str1) == 0:  # 모든 태그를 다 제거하였다면
            break  # 반복 종료
    return result  # 순수 문자열을 저장한 리스트 result를 반환


#숫자 및 구두문자 제거
def deletePun_Num(pathList):
    i = 0
    while i < len(pathList):
        pathList[i] = pathList[i].translate(str.maketrans('', '', string.digits))
        pathList[i] = pathList[i].translate(str.maketrans('', '', string.punctuation))
        pathList[i] = pathList[i].strip()
        i += 1
    return pathList


# 리스트 사전화
def makeDitionary(pathList):
    result = {}
    for i in pathList:
        result[i] = result.get(i, 0) + 1
    return result


# 메인
urlList = ['http://cse.koreatech.ac.kr',
           'https://www.koreatech.ac.kr',
           'http://www.naver.com',
           'http://www.daum.net',
           'http://www.nytimes.com']

for url in urlList:
    # html 문자열 저장
    req = requests.get(url)  # url의 html을 불러와서 req에 대입
    source = req.text        # 텍스트 화하여 source에 대입
    source = deleteStrinExEn(source)

    # 리스트 제작
    print("url : [ {0} ]".format(url))
    result_List = deletePun_Num(print_string(source))
    result_List = list(filter(None, result_List))  #리스트 내의 빈 공백 삭제
    result_List = ' '.join(result_List).split()
    result_Dict = makeDitionary(result_List)
    
    # print(result)   # 리스트 출력
    print('\n단어의 개수 : ', str(sum(result_Dict.values())))
    print(result_Dict)  #빈도 수 사전화
    print("\n\n")
```
